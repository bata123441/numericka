## Formule:
Lagrangeov interpolacijski polinom:
$${\omega _j}\left( x \right) = \prod\limits_{j = 0,j \ne i}^n {\left( {x - {x_j}} \right)} $$
$${p_L}\left( x \right) = \sum\limits_{i = 0}^n {{y_i}\frac{{{\omega _i}\left( x \right)}}{{{\omega _i}\left( {{x_i}} \right)}}} $$
Newtonov interpolacijski polinom:
$$p\left[ {{x_0}, \ldots ,{x_i}} \right] = \frac{{p\left[ {{x_1}, \ldots ,{x_i}} \right] - p\left[ {{x_0}, \ldots ,{x_{i - 1}}} \right]}}{{{x_i} - {x_0}}}$$
$$p\left[ {{x_i}} \right] = p\left( {{x_i}} \right) = {y_i}$$
$${p_N}\left( x \right) = \sum\limits_{i = 0}^n {p\left[ {{x_0}, \ldots ,{x_i}} \right]\prod\limits_{j = 0}^{i - 1} {\left( {x - {x_j}} \right)} } $$
Dekompozicija na singularne vjerojatnosti:
$$A=UDV^T$$
Generalizirani inverz:
$$A^+=VD^+U^T$$
Metoda bisekcije:
$$n \geqslant {\log _2}\left( {\frac{{b - a}}{\varepsilon }} \right) - 1$$
Newtonova metoda:
$$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$$
Aproksimacija u $L_2([a,b])$ funkcije $f(t)$ funkcijom $\varphi \left( t \right) = \sum\limits_{i = 0}^n {{c_i}{\varphi _i}\left( t \right)} $:
$$\left[ {\begin{array}{c}
  {\left\langle {{\varphi _0}|{\varphi _0}} \right\rangle }&{\left\langle {{\varphi _1}|{\varphi _0}} \right\rangle }& \cdots &{\left\langle {{\varphi _n}|{\varphi _0}} \right\rangle } \\ 
  {\left\langle {{\varphi _0}|{\varphi _1}} \right\rangle }&{\left\langle {{\varphi _1}|{\varphi _1}} \right\rangle }& \cdots &{\left\langle {{\varphi _n}|{\varphi _1}} \right\rangle } \\ 
   \vdots & \vdots & \ddots &{} \\ 
  {\left\langle {{\varphi _0}|{\varphi _n}} \right\rangle }&{\left\langle {{\varphi _1}|{\varphi _n}} \right\rangle }& \cdots &{\left\langle {{\varphi _n}|{\varphi _n}} \right\rangle } 
\end{array}} \right]\left[ {\begin{array}{c}
  {{c_0}} \\ 
  {{c_1}} \\ 
   \vdots  \\ 
  {{c_n}} 
\end{array}} \right] = \left( {n + 1} \right)\left[ {\begin{array}{c}
  {\left\langle {f|{\varphi _0}} \right\rangle } \\ 
  {\left\langle {f|{\varphi _1}} \right\rangle } \\ 
   \vdots  \\ 
  {\left\langle {f|{\varphi _n}} \right\rangle } 
\end{array}} \right]$$
Legendreovi polinomi i relacija ortogonalnosti na $L_2([-1,1])$: 
$${P_0}\left( x \right) = 1$$
$${P_1}\left( x \right) = x$$
$${P_{n + 1}}\left( x \right) = \frac{{2n + 1}}{{n + 1}}x{P_n}\left( x \right) - \frac{n}{{n + 1}}{P_{n - 1}}\left( x \right)$$
$$\left\langle {{P_n}|{P_m}} \right\rangle  = \int_{ - 1}^1 {{P_n}\left( t \right){P_m}\left( t \right)dt}  = \left\{ \begin{gathered}
  0 & ,m \ne n \\
  \frac{2}{{2n + 1}} & ,m = n \\ 
\end{gathered}  \right.$$
Čebiševljevi polinomi prve vrste i relacija ortogonalnosti na $L_2([-1,1])$:
$$\begin{gathered}
  {T_0}\left( x \right) = 1 \\ 
  {T_1}\left( x \right) = x \\ 
  {T_{n + 1}}\left( x \right) = 2x{T_n}\left( x \right) - {T_{n - 1}}\left( x \right) \\ 
\end{gathered} $$ 
$$\left\langle {{T_n}|{T_m}} \right\rangle  = \int_{ - 1}^1 {{T_n}\left( t \right){T_m}\left( t \right)\frac{1}{{\sqrt {1 - {t^2}} }}dt}  = \left\{ \begin{gathered}
  0 & ,m \ne n \\
  \pi  & ,m = n = 0 \\
  \frac{\pi }{2} & ,m = n \ne 0 \\ 
\end{gathered}  \right.$$
### Teorem: 
Među svim polinomima na $L_\infty([-1,1])$ stupnja $n$ s vodećim koeficijentom jednakim 1, polinom ${{2^{1 - n}}{T_n}}$ ima najmanju normu, odnosno ${\left\| {{2^{1 - n}}{T_n}} \right\|_\infty } = {2^{1 - n}}$ ima najmanju vrijednost.

Čebiševljevi čvorovi na $[-1,1]$:
$${x_k} = \cos \left( {\frac{{2k + 1}}{{2n}}\pi } \right),k = 0, \ldots ,n - 1$$
Linearna regresija metoda najmanjih kvadrata:
$$\overline x  = \frac{1}{n}\sum\limits_{i = 1}^n {{x_i}} ,\overline y  = \frac{1}{n}\sum\limits_{i = 1}^n {{y_i}} $$
$${S_{xy}} = \sum\limits_{i = 1}^n {{x_i}{y_i}}  - n\overline x  \cdot \overline y $$
$$b = \frac{{{S_{xy}}}}{{{S_{xx}}}},a = \overline y  - b\overline x $$
Linearna regresija QR dekompozicija:
$${R_0}x = {\left( {{Q^{\left( 1 \right)}}} \right)^T}b$$
Linearna regresija SVD:
$$x = V{\Sigma ^{+}}{U^T}b$$
Produljena trapezna formula:
$$h = \frac{{b - a}}{n},{x_i} = a + i \cdot h$$
$$\int_a^b {f\left( x \right)dx}  \approx h\left( {\frac{{f\left( a \right) + f\left( b \right)}}{2} + \sum\limits_{i = 1}^{n - 1} {f\left( {{x_i}} \right)} } \right)$$

Odredite rješenje jednadžbe $f_1(x)=f_2(x)$ na $[0,2]$ metodom bisekcije uz točnost $\epsilon=0.15$.

### <span style="color:green">Zadatak 7.b):</span>
Kodom provjerite rješenje prethodnog zadatka tako da ispišete razliku rješenja koje ste dobili i izlaza funkcije Bisekcija.

```python
def Bisekcija(f: Callable[[float], float], a: float, b: float, epsilon: float) -> float:
    '''Za danu neprekidnu funkciju f i segment [a,b] vraća nultočku od f na [a, b] uz grešku manju od epsilon'''
    n = math.ceil(math.log2((b-a)/epsilon) - 1)
    x_n = (a+b)/2
    for _ in range(n):
        if f(x_n) == 0:
            return x_n
        elif f(a) * f(x_n) < 0:
            b = x_n
        else:
            a = x_n
        x_n = (a+b)/2
    return x_n
```

Odredite rješenje jednadžbe $f_1(x)=f_2(x)$ na $[0,2]$ Newtonovom metodom uz točnost $\epsilon=0.01$.

### <span style="color:green">Zadatak 8.b):</span>
Kodom provjerite rješenje prethodnog zadatka tako da ispišete razliku rješenja koje ste dobili i izlaza vrijednosti posljednje iteracije funkcije Newton.

```python
def Newton(f: Callable[[float], float], df: Callable[[float], float], x0: float, epsilon: float) -> tuple[int, float]:
    '''Za danu funkciju f i njenu derivaciju df provodi iteracije Newtonove metode počevši iz x0 
    sve dok udaljenost vrijednosti dobivenih uzastopnim iteracijama nije manja ili jednaka od epsilon
    ili dok ne dođemo do 1000. iteracije

    Vraća tuple broja koraka i vrijednosti posljednje iteracije'''
    x1 = x0 - f(x0)/df(x0)
    i = 1
    while abs(x1 - x0) > epsilon and i < 1000:
        i += 1
        x0 = x1
        x1 = x0 - f(x0)/df(x0)
    if i == 1000:
        print("Newtonova metoda nije iskonvergirala")
    return i, x1
```

### <span style="color:green">Zadatak 9.b):</span>
Provjerite rješenje koristeći QR dekompoziciju i SVD. Nacrtajte regresijski pravac i podatke na istom grafu.

```python
A = np.array([
    [167, 1],
    [192, 1],
    [183, 1],
    [211, 1],
    [168, 1]
])
b = np.array([
    [72],
    [86],
    [77],
    [92],
    [72]
])
Q, R = np.linalg.qr(A)
x_qr = np.linalg.solve(R, Q.T @ b)
U, s, Vt = np.linalg.svd(A)
x_svd = Vt.T @ sp.linalg.diagsvd(s**(-1), 2, 5) @ U.T @ b
print("b = ", round(x_qr[0, 0], 2), "a = ", round(x_qr[1, 0], 2))
print("b = ", round(x_svd[0, 0], 2), "a = ", round(x_svd[1, 0], 2))
```

```python
x = A[:, 0]
y = b.flatten()
a, b = x_qr.flatten()
x_lin = np.linspace(x.min(), x.max(), 100)
plt.scatter(x, y)
plt.plot(x_lin, x_lin*a+b, "r")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show
```
